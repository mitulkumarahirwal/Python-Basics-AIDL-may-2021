```
# Numpy array operations
import numpy as np
a = np.array([[1, 2],[3, 4]])
b = np.array([[0, 1],[2, 2]])
print("a=\n", a, "\nb=\n", b)
```

```
#1. Element wise addition 
c = np.add(a,b)
print("c=\n",c)
```

```
#2. Element wise subtraction 
c = np.subtract(a,b)
print("c=\n",c)
```

```
#3. Element wise divide
c = np.divide(b,a)
print("c=\n",c)
```

```
#4. Element wise multiply
c = np.multiply(a,b)
print("c=\n",c)
```

```
#5. Matrix multiply
c = np.dot(a,b)
print("c=\n",c)
```

```
#6. All elements sum
s = np.sum(a)
print("s=\n",s)

# colum wise sum
c = np.sum(a,axis=0)
print("\nc=\n",c)

# row wise sum
r = np.sum(a,axis=1)
print("\nr=\n",r)
```
