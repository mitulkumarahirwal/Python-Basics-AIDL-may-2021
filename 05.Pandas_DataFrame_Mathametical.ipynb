```
import pandas as pd
# count, sum, min, max Operations on DataFrame
# DataFrame from Dictionary 

df = pd.DataFrame({'Name':['amit', 'sumit', 'riya', 'dev', 'kavita'],
                    'Age': [20, 21, 33, 23, 16],
                    'Address': ['bhopal', 'patna', 'delhi', 'bangalore', 'raipur'],
                    'kg': [56, 50, 45, 55, 66]})
df
```

```
#1. count 
df[['Age']].count()
```

```
#2. sum
df[['Age']].sum()
```

```
#3. mean
df[['Age']].mean()
```

```
#4. average
df[['Age']].sum()/df[['Age']].count()
```

```
#5. Min
df[['Age']].min()
```

```
#6. Max
df[['Age']].max()
```

```
#7. Min Index
df[['Age']].idxmin()
```

```
#8. Max Index
df[['Age']].idxmax()
```

```
#9. Summary of data
df[['Age']].describe()
```
